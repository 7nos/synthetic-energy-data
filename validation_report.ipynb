```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport 
```

```python
df = pd.read_csv("D:\kaggle_dataset\synthetic_energy_data.csv")
```

```python
print("Average Daily kWh:", df["daily_kwh"].mean())
print("Solar Adoption Rate:", df["has_solar_panel"].mean())
```

```python
real_avg_kwh = 29.0
synth_avg_kwh = df["daily_kwh"].mean()
deviation = (synth_avg_kwh - real_avg_kwh) / real_avg_kwh * 100
print(f"Deviation: {deviation:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["daily_kwh"], kde=True)
plt.axvline(real_avg_kwh, color='red', linestyle='--', label="Real-World Avg")
plt.title("Energy Consumption Distribution")
plt.legend()
plt.show()
```
